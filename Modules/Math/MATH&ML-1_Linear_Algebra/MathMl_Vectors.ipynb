```python
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [-1, 1],
    [3, 0]
])

det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_AB = det_A * det_B
print(np.around(det_AB))

print('-----')

AB = A @ B
det_AB2 = np.linalg.det(AB)
print(np.around(det_AB2))
```

```python
A = np.array([
    [1, 2],
    [3, 4]
])

inv_A = np.linalg.inv(A)

det_A = np.linalg.det(A)
det_inv_A = np.linalg.det(inv_A)

print(np.around(det_A, 1))
print('------')
print(np.around(det_inv_A, 1))
print('------')
print(np.around(det_A * det_inv_A))
```
