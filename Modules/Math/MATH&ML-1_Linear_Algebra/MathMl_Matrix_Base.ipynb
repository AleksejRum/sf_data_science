### TASK 5.1

```python
import numpy as np

Husband_Income = np.array([100,220,140])
Wife_Income = np.array([150,200,130])
Mother_In_Law_Income = np.array([90,80,100])

Husband_Consumption = np.array([50,50,60])
Wife_Consumption = np.array([100,80,140])
Mother_In_Law_Consumption = np.array([100,20,140])
```

```python
Inc = np.array([
    Husband_Income,
    Wife_Income,
    Mother_In_Law_Income
])
Inc = Inc.T

print(Inc)
print('---------')
print(Inc[0,0],Inc[1,1],Inc[2,2])
```

### TASK 5.2

```python
Cons = np.array([
    Husband_Consumption,
    Wife_Consumption,
    Mother_In_Law_Consumption
])
Cons = Cons.T

print(Cons)
print('---------')
print(Cons[0,0],Cons[1,1],Cons[2,2])
```

### TASK 5.3

```python
Inc_after_tax = Inc*0.87
print(Inc_after_tax)
print('------------')
print(Inc_after_tax[0])
```

### TASK 5.4

```python
P = Inc_after_tax - Cons
print(P)
print('----------')
print(P[2])
```
