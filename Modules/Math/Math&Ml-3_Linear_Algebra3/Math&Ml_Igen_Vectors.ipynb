```python
from sklearn import datasets 
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# загружаем датасет
from sklearn.datasets import fetch_california_housing
boston = fetch_california_housing()

#print(boston['DESCR'])
boston_data = pd.DataFrame(
    data=boston.data, #данные
    columns=boston.feature_names #наименования столбцов
)
boston_data['PRICE']=boston.target
boston_data.head()
```

```python
A = boston_data.drop('PRICE', axis=1)
y = boston_data[['PRICE']]
```

```python
C = A.corr()
fig = plt.figure(figsize=(10, 5))
sns.heatmap(C, annot=True, cmap='coolwarm');
```

```python
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(C)
print('Собственные числа: \n', eig_values)
#print('Собственные векторы: \n', eig_vectors)
```

```python
# считаем матрицу Грамма L^T*L:
print(np.round(eig_vectors.T @ eig_vectors, 2))
```

```python
x1 = np.array([1,2,1,1]).T
x2 = np.array([70,130,65,60]).T

C = np.corrcoef(x1, x2)
C
```

```python
eig_values, eig_vectors = np.linalg.eig(C)
display(eig_values)
display(eig_vectors)
```

```python
x1_norm = (x1-x1.mean())/np.linalg.norm(x1)
x2_norm = (x2-x2.mean())/np.linalg.norm(x2)
```

```python
x_main = x1_norm * eig_vectors[0][0] + x2_norm * eig_vectors[1][0]
x_main_norm = (x_main - x_main.mean())/np.linalg.norm(x_main)
```

```python
display(x_main_norm)
```

### SVD

```python
A = np.array([
    [2, 6],
    [5, 3],
    [-4, 0]
])

AT_A = A.T @ A
AT_A
```

```python
eig_values, eig_vectors = np.linalg.eig(AT_A)
display(eig_values)
display(eig_vectors.round(1))
```

```python
A_AT = A @ A.T
A_AT
```

```python
eig_values2, eig_vectors2 = np.linalg.eig(A_AT)
display(eig_values2.round())
display(eig_vectors2.round(1))
```
