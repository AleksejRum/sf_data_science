### UNION  (C = A ⋃ B)

```python
A = {1, 2, 3, 4}
B = {2, 3, 5}

C = A.union(B)
print(C)
```

### INTERSECTION (C = A ∩ B)

```python
A = {1, 2, 3, 4}
B = {2, 3, 5}

C = A.intersection(B)
print(C)
```

### DIFFERENCE (C = A \ B)

```python
A = {1, 2, 3, 4}
B = {2, 3, 5}

C = A.difference(B)
print(C)
```

### SYMMETRIC DIFFERENCE (C = A Δ B)

```python
A = {1, 2, 3, 4}
B = {2, 3, 5}

C = A.symmetric_difference(B)
print(C)
```

## Eigenvectors and numbers

```python
# TASK 3.2

import numpy as np
A = np.array([
    [-7, 6],
    [-12, 11]
])

λ1 = -1
λ2 = 5

v1 = np.array([1,2])
v2 = np.array([1,1])

print(A @ v1)
print(A @ v2)
print('------')
print(v1 * λ1)
print(v2 * λ2)
print('-----')
print(v1 * λ2)
print(v2 * λ1)
```

```python
# TASK 3.3

A = np.array([
    [5, 1],
    [1, 5]
])

λ1 = 6
λ2 = 4

v1 = np.array([1,1])
v2 = np.array([-1,1])
v3 = np.array([0,2])
v4 = np.array([4,4])

print(A @ v1)
print(A @ v2)
print(A @ v3)
print(A @ v4)
print('------')
print(v1 * λ1)
print(v2 * λ1)
print(v3 * λ1)
print(v4 * λ1)
print('------')
print(v1 * λ2)
print(v2 * λ2)
print(v3 * λ2)
print(v4 * λ2)
```
